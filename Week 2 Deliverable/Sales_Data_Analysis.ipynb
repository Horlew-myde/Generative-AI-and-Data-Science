```python
#Load Libraries and Data

import pandas as pd
import matplotlib.pyplot as plt

# # The URL where the sales data is stored
# data_url = 'https://pynative.com/wp-content/uploads/2019/01/company_sales_data.csv'

# # Read the data into a pandas DataFrame
# sales_data = pd.read_csv(data_url)

# Load the data
sales_data = pd.read_csv('company_sales_data.csv')


# Display the first few rows to check the data
print("Data loaded successfully. Here are the first 5 rows:")
sales_data.head()
```

```python
#Exercise 1 - Plot Total Profit

print("--- Chart 1: Company's Total Profit per Month ---")

# Set the size of the plot
plt.figure(figsize=(10, 6))

# Plot the 'month_number' on the x-axis and 'total_profit' on the y-axis
plt.plot(
    sales_data['month_number'],
    sales_data['total_profit'],
    marker='o', # Puts a dot on each month's data point
    linestyle='-',
    color='purple'
)

# Add titles and labels for clarity
plt.title("Company's Total Profit per Month")
plt.xlabel("Month Number")
plt.ylabel("Profit in Dollars ($)")

# Make sure every month number is shown on the x-axis
plt.xticks(sales_data['month_number'])

# Add a grid for easier reading
plt.grid(True, linestyle='--')

# Display the final plot
plt.show()
```

```python
#Exercise 2 - Subplots for Soap and Facewash Sales

print("--- Chart 2: Comparing Bathing Soap and Facewash Sales ---")

# Create a figure and a set of two subplots, arranged in 2 rows, 1 column
fig, (plot1, plot2) = plt.subplots(2, 1, figsize=(10, 8))

# --- Top Plot: Bathing Soap ---
plot1.plot(
    sales_data['month_number'],
    sales_data['bathingsoap'],
    marker='s', # Use a square marker
    color='green'
)
plot1.set_title("Bathing Soap Sales Per Month")
plot1.set_ylabel("Units Sold")
plot1.grid(True, linestyle=':') # Dotted grid style

# --- Bottom Plot: Facewash ---
plot2.plot(
    sales_data['month_number'],
    sales_data['facewash'],
    marker='^', # Use a triangle marker
    color='orange'
)
plot2.set_title("Facewash Sales Per Month")
plot2.set_xlabel("Month Number")
plot2.set_ylabel("Units Sold")
plot2.grid(True, linestyle=':')

# Make sure every month number is shown on the shared x-axis
plt.xticks(sales_data['month_number'])

# Adjust the layout to prevent titles/labels from overlapping
plt.tight_layout()

# Display the final plots
plt.show()
```
